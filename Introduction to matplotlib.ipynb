# pyplot vs matplotlib

pyplot-->when plotting something quick
OO method-->when plotting something more advanced

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50
```

```python
#pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,6));
```

```python
#OO Method
fix, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
#ax.set_xlim([45,100]);
```

# Ploting from pandas dataframe 6

```python
## OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])

#customise the plot
ax.set(title="Heart disease and choloestrol levels",
       xlabel="age",
       ylabel="cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

# plotting from pandas dataframe 7

```python
over_50.head()
```

```python
#subplots of chol, age,thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))

#add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                  c=over_50["target"])

# customize ax0
ax0.set(title="Heart disease and choloestrol levels",
       xlabel="age",
       ylabel="cholesterol")

#add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");

# add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                  c=over_50["target"])

# customize ax1
ax1.set(title="Heart disease and max heart rate",
       xlabel="age",
       ylabel="max heart rate")

#add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");

# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
                    
```

# Customizing plots 2

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50
```

```python
#set the style
plt.style.use("seaborn-whitegrid")

## OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"],
                  cmap="winter") #this change colour scheme
                                #matplotlib colours tutorial

#customise the plot
ax.set(title="Heart disease and choloestrol levels",
       xlabel="age",
       ylabel="cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
#customising the x and y limitations

#subplots of chol, age,thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))

#add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                  c=over_50["target"],
                   cmap="winter")

# customize ax0
ax0.set(title="Heart disease and choloestrol levels",
       xlabel="age",
       ylabel="cholesterol")
#change the x-axis limit
ax0.set_xlim([50,80])
#add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");

# add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter") #for the colour of the dots

# customize ax1
ax1.set(title="Heart disease and max heart rate",
       xlabel="age",
       ylabel="max heart rate")
#change the x-axis limit(to remove line in corner)
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");

# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
                    
```
